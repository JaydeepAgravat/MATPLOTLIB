## 2D Line plot

- Bivariate Analysis
- categorical -> numerical and numerical -> numerical
- Use case - Time series data

```python
batsman = pd.read_csv('/content/drive/MyDrive/MATPLOTLIB/DATASET/sharma-kohli.csv')
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dashdot',linewidth=2,marker='o',label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc='upper left')
# plt.grid()
plt.show()
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.show()
```

## Subplots

```python
bt.head()
```

```python
plt.figure(figsize=(15,6))
plt.scatter(bt['avg'],bt['strike_rate'])
plt.title('Something')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')

plt.show()
```

```python
fig,ax = plt.subplots(figsize=(15,6))

ax.scatter(bt['avg'],bt['strike_rate'],color='red',marker='+')
ax.set_title('Something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')

fig.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))

ax[0].scatter(bt['avg'],bt['strike_rate'],color='red')
ax[1].scatter(bt['avg'],bt['runs'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')


ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')

plt.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(bt['avg'],bt['strike_rate'],color='red')
ax[0,1].scatter(bt['avg'],bt['runs'])
ax[1,0].hist(bt['avg'])
ax[1,1].hist(bt['runs'])
plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(bt['avg'],bt['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2)
ax2.hist(bt['runs'])

ax3 = fig.add_subplot(2,2,3)
ax3.hist(bt['avg'])

plt.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,sharex=True,figsize=(10,10))
ax[1,1].hist(bt['runs'])
plt.show()
```

## 3D Line Plot

```python
x = [0,1,5,25]
y = [0,10,13,0]
z = [0,13,20,9]

fig = plt.figure()

ax = plt.subplot(projection='3d')

ax.scatter3D(x,y,z,s=[100,100,100,100])
ax.plot3D(x,y,z,color='red')

plt.show()
```

## 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
plt.show()
```

## Contour Plots

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
plt.show()
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
plt.show()
```
